```python
# 18

# see those links clrly 

#ex op - [7,4,1],[8,5,2],[9,6,3]] --> inplace

m=[[1,2,3],[4,5,6],[7,8,9]]
n=len(m)

def swap(a,b):
    t=a
    a=b
    b=t
    return a,b

#transpose of matrix
for i in range(0,n):
    for j in range(i,n):
        m[i][j],m[j][i]=swap(m[i][j],m[j][i])       
print("step 1 is to transpose matrix - after transpose\n",m)

#now swap each row elements 
for i in range(n):
    l=0
    h=n-1
    #print("i is -",i)
    while(l<h):
        #print(m[i][l],m[i][h])
        m[i][l],m[i][h]=swap(m[i][l],m[i][h])
        l+=1
        h-=1
    #print("i th m is --",m)
    
        
print("\n\nstep 2 is to swap ele in a single row wise- final ans\n",m)
```
